```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from IPython.display import display

# Assuming 'mcdonalds' is your DataFrame and MD_x is prepared as before
MD_x = mcdonalds.iloc[:, :11].values  # Select first 11 columns as NumPy array
MD_x = (MD_x == "Yes").astype(int)  # Convert "Yes" to 1, others to 0

# --- PCA ---
MD_pca = PCA()
MD_pca.fit(MD_x)

# --- Variance Explained ---
variance_explained = MD_pca.explained_variance_ratio_
cumulative_variance = np.cumsum(variance_explained)

# --- Visualization ---
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(variance_explained) + 1), variance_explained, marker='o', linestyle='--')
plt.title('Scree Plot: Explained Variance by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.xticks(range(1, len(variance_explained) + 1))
plt.grid(True)
plt.show()

# --- Summary Table ---
summary_table = pd.DataFrame({
    'Standard deviation': np.sqrt(MD_pca.explained_variance_),
    'Proportion of Variance': variance_explained,
    'Cumulative Proportion': cumulative_variance
}, index=[f'PC{i}' for i in range(1, len(variance_explained) + 1)])
display(summary_table)

# --- Loadings (Component Coefficients) ---
loadings_table = pd.DataFrame(MD_pca.components_.T, columns=[f'PC{i}' for i in range(1, MD_pca.n_components_ + 1)],
                              index=mcdonalds.columns[:11])
display(loadings_table)
```

**Using Mixtures of Distributions**

The mixture model maximises the likelihood to extract segments (as opposed to
minimising squared Euclidean distance, as is the case for k-means). The call to
stepFlexmix() extracts two to eight segments (k = 2:8) using ten random
restarts of the EM algorithm (nrep), model = FLXMCmvbinary() for a
segment-specific model consisting of independent binary distributions and no
intermediate output about progress (verbose = FALSE).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
from sklearn.mixture import GaussianMixture  # For mixture modeling

# Assuming 'mcdonalds' is your DataFrame and MD_x is prepared as before
MD_x = mcdonalds.iloc[:, :11].values  # Select first 11 columns as NumPy array
MD_x = (MD_x == "Yes").astype(int)  # Convert "Yes" to 1, others to 0

# --- K-means Clustering ---
def kmeans_clustering(data, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=1234)
    kmeans.fit(data)
    return kmeans.labels_

# --- K-means Scree Plot ---
n_clusters_range = range(2, 9)
kmeans_inertia = []
for n_clusters in n_clusters_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=1234)
    kmeans.fit(MD_x)
    kmeans_inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 6))
plt.plot(n_clusters_range, kmeans_inertia, marker='o', linestyle='--')
plt.title('K-means Scree Plot')
plt.xlabel('Number of Segments')
plt.ylabel('Sum of Within-Cluster Distances')
plt.xticks(n_clusters_range)
plt.grid(True)
plt.show()

# --- Global Stability Boxplot ---
def bootstrap_evaluation(data, n_clusters_range, n_bootstraps=100):
    bootstrap_ari_scores = []
    for n_clusters in n_clusters_range:
        ari_scores = []
        for _ in range(n_bootstraps):
            bootstrap_sample = resample(data, random_state=1234)
            cluster_labels = kmeans_clustering(bootstrap_sample, n_clusters)
            ari_scores.append(adjusted_rand_score(cluster_labels, cluster_labels))
        bootstrap_ari_scores.append(ari_scores)
    return bootstrap_ari_scores

bootstrap_ari_scores = bootstrap_evaluation(MD_x, n_clusters_range)

plt.figure(figsize=(8, 6))
plt.boxplot(bootstrap_ari_scores, labels=n_clusters_range)
plt.title('Global Stability Boxplot')
plt.xlabel('Number of Segments')
plt.ylabel('Adjusted Rand Index')
plt.grid(True)
plt.show()

# --- Mixture Modeling ---
def mixture_modeling(data, n_components):
    gmm = GaussianMixture(n_components=n_components, random_state=1234)
    gmm.fit(data)
    return gmm.predict(data)

# Assuming MD.k4 is your selected k-means solution with 4 clusters
MD_k4 = kmeans_clustering(MD_x, 4)

# --- Mixture Model with 4 Components (Similar to MD.m4 in R) ---
MD_m4 = mixture_modeling(MD_x, 4)

# --- Cross-tabulation (Similar to table() in R) ---
from collections import Counter

def cross_tabulation(labels1, labels2):
    cross_tab = Counter(zip(labels1, labels2))
    labels1_unique = sorted(set(labels1))
    labels2_unique = sorted(set(labels2))
    table_data = [[cross_tab[(l1, l2)] for l2 in labels2_unique] for l1 in labels1_unique]
    return pd.DataFrame(table_data, index=labels1_unique, columns=labels2_unique)

cross_tab = cross_tabulation(MD_k4, MD_m4)
print("Cross-tabulation of K-means and Mixture Model Clusters:")
print(cross_tab)

# --- Segment Level Stability Plot ---
# Assuming MD.k4 is your selected k-means solution with 4 clusters
# MD.k4 = kmeans_clustering(MD_x, 4)  # Replace with your actual solution

# Replace this with your actual stability calculation for each segment
segment_stability = [0.8, 0.9, 0.7, 0.85]  # Dummy values for demonstration

plt.figure(figsize=(8, 6))
plt.bar(range(1, 5), segment_stability)
plt.title('Segment Level Stability Plot (4 Segments)')
plt.xlabel('Segment Number')
plt.ylabel('Segment Stability')
plt.xticks(range(1, 5))
plt.ylim(0, 1)
plt.grid(True)
plt.show()

# --- Summary Table ---
summary_table = pd.DataFrame({
    'Segment': range(1, 5),
    'Stability': segment_stability
})
print(summary_table)

# --- Probability Tables ---
def segment_probabilities(data, cluster_labels):
    n_clusters = len(set(cluster_labels))
    segment_probs = []
    for cluster in range(n_clusters):
        segment_data = data[cluster_labels == cluster]
        probs = segment_data.mean(axis=0)
        segment_probs.append(probs)
    return pd.DataFrame(segment_probs, index=range(n_clusters), columns=mcdonalds.columns[:11])

kmeans_probs = segment_probabilities(MD_x, MD_k4)
mixture_probs = segment_probabilities(MD_x, MD_m4)

print("\nK-means Segment Probabilities:")
print(kmeans_probs)

print("\nMixture Model Segment Probabilities:")
print(mixture_probs)
```

**Using Mixtures of Regression Models**

1. Objective: Find market segments where love or hate for McDonalds is driven by similar perceptions.

Approach: Use finite mixtures of linear regression models (latent class regressions).

2. Variables:
Dependent Variable (y): Degree of love or hate for McDonalds, measured by responses to "I LIKE MCDONALD'S" on an 11-point scale (1 = I LOVE IT!, 11 = I HATE IT!).

Independent Variables (x): Perceptions of McDonalds.

3. Segmentation:

Focuses on unobserved segmentation variables (regression coefficients).

Market segments consist of consumers with similar perception effects on their liking of McDonalds.

4. Data Transformation:

Convert the ordinal dependent variable "LIKE" to a numeric scale.

Numeric codes: 1 (I LOVE IT!) to 11 (I HATE IT!).

Compute the new numeric variable as 6 minus the numeric code.

**Exploring the McDonald's Dataset**

This section focuses on the initial exploration of the "mcdonalds" dataset using R. The exploration involves loading the dataset, examining its structure, and understanding its key characteristics. Key steps include:

1. Loading the Dataset: Using the MSA package to access the "mcdonalds" dataset.
2. Inspecting Variable Names: Listing all the columns to understand what attributes are included.
3. Checking Dimensions: Identifying the number of observations and variables to gauge the dataset's size.
4. Previewing Data: Viewing the first few rows to get a sense of the data's content and format.
5. This step is critical for getting familiar with the dataset before conducting further analysis or modeling.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

mcdonalds = pd.read_csv('/mcdonalds.csv')


print(mcdonalds.columns)


print(mcdonalds.shape)


print(mcdonalds.head(3))


plt.figure(figsize=(8, 6))
sns.countplot(x='Like', data=mcdonalds)
plt.title('Distribution of Likes')
plt.xlabel('Like')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(8, 6))
plt.hist(mcdonalds['Age'], bins=10)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Age', data=mcdonalds)
plt.title('Age Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

Principal components analysis of the fast food data set

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Assuming 'mcdonalds' is your DataFrame
MD_x = mcdonalds.iloc[:, :11].values  # Select first 11 columns as NumPy array
MD_x = (MD_x == "Yes").astype(int)  # Convert "Yes" to 1, others to 0
col_means = MD_x.mean(axis=0).round(2)  # Calculate column means and round

# --- Visualization ---
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(range(len(col_means)), col_means)
ax.set_xticks(range(len(col_means)))
ax.set_xticklabels(mcdonalds.columns[:11], rotation=45, ha='right')
ax.set_title('Proportion of "Yes" Responses for Each Attribute')
ax.set_xlabel('Attribute')
ax.set_ylabel('Proportion')
plt.tight_layout()
plt.show()

# --- Descriptive Statistics Table ---
desc_stats = pd.DataFrame(MD_x, columns=mcdonalds.columns[:11]).describe().T
display(desc_stats)

# --- Frequency Table ---
freq_table = pd.DataFrame(col_means, columns=['Proportion of "Yes"'], index=mcdonalds.columns[:11])
display(freq_table)

print(col_means)
```
